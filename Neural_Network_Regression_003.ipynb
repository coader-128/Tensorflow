```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.losses import mean_absolute_error

# set random seed
tf.random.set_seed(42)

#1 . create a model using the Sequential

model = Sequential([
    tf.keras.layers.Dense(100 , activation = 'relu' ,input_shape = (1,)),
    tf.keras.layers.Dense(100 , activation = 'relu'),
    tf.keras.layers.Dense(100 , activation = 'relu' ),
    tf.keras.layers.Dense(100 , activation = 'relu' ),
    tf.keras.layers.Dense(100 , activation = 'relu' ),
    tf.keras.layers.Dense(1 )

])


# 2 . compile the model

model.compile( loss  = mean_absolute_error ,
              optimizer = tf.keras.optimizers.Adam(learning_rate= 0.0001),
               metrics = 'mae')

# 3 . fit the model

model.fit(X , y ,epochs = 1000)

#
```

```python
# lets create a plotting function

def plot_predictions(train_data = x_train,
                     train_labels = y_train,
                     test_data = x_test,
                     test_labels = y_test,
                     prediction = y_pred):
  plt.figure(figsize = (10,7))
  #plot training data in blue
  plt.scatter(train_data , train_labels , c = 'b' , label = 'Training data')
  # plot testing data in green
  plt.scatter(test_data , test_labels , c = 'g' , label = 'testing data')
  #plot models prediction
  plt.scatter(test_data , prediction , c = 'r' , label = 'prediction')
  plt.legend();
```

## pre processing data (Normalization and standardization)

```python
# reading the datatset
insurance =  pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance
```

## to prepare our data we need help from sklearn



```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler , OneHotEncoder
from sklearn.model_selection import train_test_split
#create a column transformer

ct = make_column_transformer(
    (MinMaxScaler() , ['age', 'bmi' , 'children']) , # turn alll values in columns between 0 to 1
    (OneHotEncoder(handle_unknown = 'ignore' ,) , ['sex' , 'smoker' , 'region'])
)


# create X and Y

x = insurance.drop('charges' , axis = 1)
y = insurance['charges']

# data splitting
x_train , x_test , y_train  ,y_test = train_test_split(x, y , test_size=0.2 , random_state = 43)

# fit the colomn transformer to the data

ct.fit(x_train)

#transform training and test with normalization and one hot encoder
x_train_normal = ct.transform(x_train)
x_test_normal = ct.transform(x_test)

```

```python
# visualisation
x_train.loc[0] ##
```

```python
x_train_normal[0]
```

```python
x_train.shape , x_train_normal.shape
```

```python
# now build neural network on it

tf.random.set_seed(43)

#model making

model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#model compile

model_4.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ['mae'])


# fit the model

model_4.fit(x_train_normal , y_train , epochs = 100 , verbose = 1 )
```

```python
# evaluate the model

model_4.evaluate(x_test_normal , y_test)
```
